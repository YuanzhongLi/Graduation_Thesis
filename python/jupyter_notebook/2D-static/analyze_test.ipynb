```python
def get_times(file_path):
    with open(file_path, 'r') as f:
        txt_lines = f.readlines()
        times = []
        for line in txt_lines:
            line_list = line.split(', ')
            for i in range(4):
                times.append(float(line_list[i*2]))
        return times
```

```python
times = get_times('test_amp_vx.txt')
```
